```python
import numpy as np
```

```python
negative = []
with open("words-negative.csv", "rb") as file:
    reader = csv.reader(file)
    for row in reader:
        negative.append(row)
```

```python
positive = []
with open("words-positive.csv", "rb") as file:
    reader = csv.reader(file)
    for row in reader:
        positive.append(row)
```

```python
def sentiment(text):
    text_sent = nltk.sent_tokenize(text)
    for sentence in text_sent:
        n_count = 0
        p_count = 0
        sent_words = nltk.word_tokenize(sentence)
        for word in sent_words:
            for item in positive:
                if(word == item[0]):
                    p_count += 1
            for item in negative:
                if(word == item[0]):
                    n_count += 1
        if(p_count > 0 and n_count == 0):
            print "+ : " + sentence
        elif(n_count%2 > 0):
            print "- : " + sentence
        elif(n_count%2 == 0 and n_count > 0):
            print "+ : " + sentence
        else:
            print "? : " + sentence
```

```python
sentiment("It was terribly bad")
```

```python
sentiment("Actually, it was not bad at all.")
```

```python
sentiment("This is a sentence about nothing.")
```

```python
def sentiment2(text):
    temp = []
    text_sent = nltk.sent_tokenize(text)
    for sentence in text_sent:
        n_count = 0
        p_count = 0
        sent_words = nltk.word_tokenize(sentence)
        for word in sent_words:
            for item in positive:
                if(word == item[0]):
                    p_count += 1
            for item in negative:
                if(word == item[0]):
                    n_count += 1
        if(p_count > 0 and n_count == 0):
            temp.append(1)
        elif(n_count%2 > 0):
            temp.append(-1)
        elif(n_count%2 == 0 and n_count > 0):
            temp.append(1)
        else:
            temp.append(0)
    return temp
```

```python
comments = []
with open("reviews.csv", "rb") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
```
