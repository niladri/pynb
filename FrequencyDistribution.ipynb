```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```python
len(alice)
```

```python
alice_fd = nltk.FreqDist(alice)
```

```python
alice_fd
```

```python
alice_fd["rabbit"]
```

```python
alice_fd["Rabbit"]
```

```python
alice_fd.most_common()
```

```python
alice_fd.most_common(15)
```

```python
alice_fd.hapaxes()
```
