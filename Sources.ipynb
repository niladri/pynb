```python
from urllib import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
```

```python
url
```

```python
html = urlopen(url).read()
```

```python
import urllib.request as ur
```

```python
req = ur.Request(url)
```

```python
with ur.urlopen(req) as response:
    html = response.read()
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
web_str = BeautifulSoup(html).get_text()
```

```python
web_str = BeautifulSoup(html, "lxml").get_text()
```

```python
web_str
```

```python
start = web_str.find("Python is a widely used high-level programming language")
```

```python
start
```

```python
end = web_str.find("CPython is managed by the non-profit Python Software Foundation.")
```

```python
end
```

```python
last_sent = len("CPython is managed by the non-profit Python Software Foundation")
```

```python
intro = web_str[start:end+last_sent]
```

```python
intro
```

```python
intro_tokens = nltk.word_tokenize(intro)
```

```python
intro_tokens
```

```python
import csv
```

```python
comments = []
with open("reviews.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
```

```python
len(comments)
```

```python
comments[0]
```

```python
tokens = [nltk.word_tokenize(str(entry)) for entry in comments]
```

```python
tokens
```
