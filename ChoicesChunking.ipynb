```python
import nltk
```

```python
bs = nltk.corpus.gutenberg.words("bryant-stories.txt")
```

```python
bs_norm = [word.lower() for word in bs if word.isalpha()]
```

```python
print len(bs)
print len(bs_norm)
```

```python
bs_tags = nltk.pos_tag(bs_norm, tagset = "universal")
```

```python
bs_cfd = nltk.ConditionalFreqDist(bs_tags)
```

```python
bs_cfd[:10]
```

```python
bs_cfd
```

```python
for ((word1, tag1), (word2, tag2), (word3, tag3)) in nltk.trigrams(bs_tags):
    if tag1 == "NOUN" and word2 == "or" and tag3 == "NOUN":
        print word1 + " " + word2 + " " + word3
```

```python
text = "I will go to the coffee shop in New York after I get off the jet plane."
```

```python
text_tag = nltk.pos_tag(nltk.word_tokenize(text))
```

```python
text_tag
```

```python
sequence = '''
Chunk:
{<NNPS>+}
{<NNP>+}
{<NN>+}'''
```

```python
NPChunker = nltk.RegexpParser(sequence)
```

```python
result = NPChunker.parse(text_tag)
```
